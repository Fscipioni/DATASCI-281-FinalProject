### Helper functions

```python
def display_cm(y_label, y_pred, laebls):

    # Compute confusion matrix
    cm = confusion_matrix(y_label, y_pred)

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels)
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title("Confusion Matrix")
    plt.show()
```

```python
print(f"Feature shape: {X_train.shape}")
print(f"Label shape: {y_train.shape}")
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.model_selection import RepeatedStratifiedKFold, GridSearchCV


#  Define your parameter grid
param_grid = {
    'svc__C': [0.1, 1, 10],
    'svc__gamma': ['scale', 'auto', 0.1, 1],
    'svc__kernel': ['rbf', 'poly', 'sigmoid']
}


# undersample with 500 samples from each class
undersample = RandomUnderSampler(sampling_strategy={0: 500, 1: 500}, random_state=42)

# Define a pipeline with undersampling
pipeline = ImbPipeline([
    ('scaler', StandardScaler()),
    ('undersample', RandomUnderSampler()),
    ('svc', SVC())
])

# Use repeated stratified k-fold
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=42)

# Set up GridSearchCV with the pipeline
grid_search = GridSearchCV(pipeline, param_grid, cv=cv, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

print(f"Best parameters: {grid_search.best_params_}")
best_svm_model = grid_search.best_estimator_
```

```python
# Best parameters: {'svc__C': 0.1, 'svc__gamma': 0.1, 'svc__kernel': 'rbf'}
```

```python
# Get a balanced subset from the test dataset to evaluate the performance

rand_sampler = RandomUnderSampler(sampling_strategy='auto', random_state=42)
X_test_balanced, y_test_balanced = rand_sampler.fit_resample(X_test, y_test)
```

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.layers import BatchNormalization
```

```python
image_dir = './Data/WithText'  
image_size = (224, 224)  


images = []
labels = []

# Iterate over each image in the directory
for filename in df_features['tconst']:

    label = np.array(df_features[df_features['tconst'] == filename]['label'])
    
    img_path = os.path.join(image_dir, filename + '.jpg')
    im = Image.open(img_path)
    image = np.array(im.convert("RGB"))
    im.close()
    
#     print(image.shape)
    
    if image is None:
        continue  # Skip corrupted or unreadable images
    
#     # Resize the image to a consistent size 
#     img_array = cv2.resize(image, image_size)
    
#     # flatten
#     img_array = img_array.flatten()
    
#     images.append(img_array)
#     labels.append(label)

    
    target_size = 224
    
    h, w, _ = image.shape
    scale = target_size / min(h, w)
    new_w, new_h = int(w * scale), int(h * scale)
    
    # Resize image
    resized = cv2.resize(image, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
    
    # Create new image with target size and paste resized image in the center
    padded = np.zeros((target_size, target_size, 3), dtype=np.uint8)
    
    # Compute cropping/padding
    x_start = max((new_w - target_size) // 2, 0)
    y_start = max((new_h - target_size) // 2, 0)

    x_end = x_start + target_size
    y_end = y_start + target_size

    # Crop resized image if needed
    cropped = resized[y_start:y_end, x_start:x_end]

    # If the resized image is smaller than target, center it in padded canvas
    h_crop, w_crop = cropped.shape[:2]
    y_off = (target_size - h_crop) // 2
    x_off = (target_size - w_crop) // 2

    padded[y_off:y_off+h_crop, x_off:x_off+w_crop] = cropped
    
    if padded.shape[0] != 224 or padded.shape[1] != 224:
        print(label)
    
    images.append(padded)
    labels.append(label)

    
    

# for path in image_paths:
#     try:
#         img = Image.open(path).convert('RGB')  # ensure 3 channels
#         img = img.resize(image_size)
#         img_array = np.array(img)
        
#         # flatten 224 * 224 * 3)
#         img_array = img_array.flatten()

#         images.append(img_array)
#         image_names.append(os.path.basename(path))

#     except Exception as e:
#         print(f"Error loading {path}: {e}")


# df_images = pd.DataFrame(images)
# df_images.insert(0, 'filename', image_names)  # optional: keep track of file names

# print(df_images.shape)
# df_images.head()
```
